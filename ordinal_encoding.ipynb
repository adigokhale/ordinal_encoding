```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("customer.csv")
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
# ordinal encoding ordinaln review and education
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
# give less value to Poor than Average
# give less value to School than UG
```

```python
# fit on training
oe.fit(X_train)
```

```python
# transform on training and testing
X_train_transformed = oe.transform(X_train)
X_test_transformed = oe.transform(X_test)
```

```python
X_train_transformed = pd.DataFrame(X_train_transformed,columns=X_train.columns)
```

```python
X_test_transformed = pd.DataFrame(X_test_transformed,columns=X_test.columns)
```
